```python
#Step_1 :- asking for the names of the players.

player_1 = input("Enter the Player_1's name:-")  
player_2 = input("Enter the Player_2's name:-")
```

```python
print("Player 1 =",player_1)
print("Player 2 =",player_2)
```

```python
#Step_2 :- assigning the respective markers to the players

def player_input(): 
    marker = ''
    while marker != 'X' and marker !='O':
        marker = input("Player 1, choose X or O: ").upper()
    Player1 = marker
    if Player1 =='X':
        Player2 ='O'
    else:
        Player2 = 'X'
    return (Player1,Player2)
```

```python
#Step3 : Creating the display for tic tac toe board.

def display_board(board):   
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
    
```

```python
#Step 4 : Assigning the values of board to the respective markers.

def place_marker(board, marker, position):
    
    board[position] = marker;   # now board with its respective position get the marker.
    
```

```python
#Step 5 : Returning the all the posibility where the player can win the game.

def win_func(board, mark):
    
    return ((board[1] == mark and board[2] == mark and board[3] == mark) or 
    (board[4] == mark and board[5] == mark and  board[6] == mark) or
    ( board[7] == mark and board[8]== mark and board[9] == mark ) or
     (board[1] == mark and board[4]== mark and board[7] == mark)  or 
    (board[2] == mark and board[5] == mark and  board[8] == mark) or
    ( board[3] == mark and board[6]== mark and board[9] == mark ) or
      (board[1] == mark and board[5] == mark and  board[9] == mark) or
    ( board[3] == mark and board[5]== mark and board[7] == mark ))
```

```python
#Step 6 : Randomly generating the players who will go first .

import random

def choose_first():
    toss = random.randint(0,1)
    if toss == 0:
        return player_1
    else:
        return player_2
    
```

```python
#Step 7 : can't reassign the same position if once filled

def space_check(board, position):
    return board[position] ==' '
```

```python
#Step 8 : Checking all the position get filled up !

def full_board_check(board):
    
    for i in range(1,10): 
        if space_check(board,i):
            return False
    
    return True
```

```python
#Step 8 : Manual Method

def manual_choice(board): 
    
    position = 0
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board,position): #if no more positions are left then loop gets terminated.
            position = int(input('Select the position : (1-9) '))
            
    return position
```
